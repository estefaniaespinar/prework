```python
#1. Robin Hood is famous for hitting an arrow with another arrow. Find the coordinates of the points where an arrow hits another arrow.

points = [(4, 5), (-0, 2), (4, 7), (1, -3), (3, -2), (4, 5), (3, 2), (5, 7), (-5, 7), (2, 2), (-4, 5), (0, -2),(-4, 7), (-1, 3), (-3, 2), (-4, -5), (-3, 2), (5, 7), (5, 7), (2, 2), (9, 9), (-8, -9)]

uniquevalues=[]
arrowhitsarrow=[]
 
for x in points:
    if x not in uniquevalues:
        uniquevalues.append(x)
    elif x in uniquevalues and x not in arrowhitsarrow:
        arrowhitsarrow.append(x)
 
print(arrowhitsarrow)

#2. Calculate how many arrows have fallen in each quadrant.¶

quadrant1= 0
quadrant2= 0
quadrant3= 0
quadrant4= 0
center=0
x_axis=0
y_axis=0

for x in points:
    if x[0] > 0 and x[1] > 0:
        quadrant1 += 1
    elif x[0] == 0 and x[1] != 0:
        x_axis += 1
    elif x[0] < 0 and x[1] > 0:
        quadrant2 += 1
    elif x[0] != 0 and x[1] == 0:
        y_axis += 1
    elif x[0] < 0 and x[1]< 0:
        quadrant3 += 1
    elif x[0] > 0 and x[1] <0:
        quadrant4 += 1
    else: 
        center += 1      
    
 
print(quadrant1)
print(quadrant2)
print(quadrant3)
print(quadrant4)
print(x_axis)
print(y_axis)
print(center)

#3. Find the point closest to the center. Calculate its distance to the center.¶

import math 
 
center=[0,0]
 
distance=[]
 
closest_center=[]
closest_distance=[]
euclidean=math.sqrt((x[0]-center[0])**2 + (x[1]-center[1])**2)
 
 
for x in points:
    distance.append(math.sqrt((x[0]-center[0])**2 + (x[1]-center[1])**2))
  
for x in points:
    if math.sqrt((x[0]-center[0])**2 + (x[1]-center[1])**2) == min(distance):
        closest_center.append(x)
 
print("The closest points to the center are:",closest_center)

for x in closest_center:
	closest_distance.append(euclidean)
print("The closest points to the center are:",closest_center, "and its distance are",closest_distance,"respectively")

#4. If the archery target has a radius of 9, calculate the number of arrows that won't hit the target.¶

import math 
out_of_target=0
for x in points:
    if math.sqrt((x[0]-center[0])**2 + (x[1]-center[1])**2) > 9:
        out_of_target += 1
print(out_of_target)

```
